```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_20_path = Path('athletic_sales_2020.csv')
athletic_sales_21_path = Path('athletic_sales_2021.csv')

athletic_sales_20_df = pd.read_csv(athletic_sales_20_path, index_col="retailer")
athletic_sales_21_df = pd.read_csv(athletic_sales_21_path, index_col="retailer")

#athletic_sales_20_df = pd.read_csv(athletic_sales_20_path, index_col='city')
#athletic_sales_21_df = pd.read_csv(athletic_sales_21_path, index_col='city')

```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

combined_sales_df = pd.concat([athletic_sales_20_df, athletic_sales_21_df], axis=0, join="inner")
reindex_combined_sales_df = combined_sales_df.reset_index(drop=False)
reindex_combined_sales_df.head()
```

```python
# Check if any values are null
null_df = reindex_combined_sales_df.isnull().values.any()
null_df
#So, there are no null values here
```

```python
# Check the data type of each column
reindex_combined_sales_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
reindex_combined_sales_df['invoice_date'] = pd.to_datetime(reindex_combined_sales_df['invoice_date'], errors='coerce')

```

```python
# Confirm that the "invoice_date" data type has been changed.
reindex_combined_sales_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".


products_sold = reindex_combined_sales_df.groupby(['region', 'state', 'city'])['units_sold'].agg(Total_Products_Sold=('sum'))
renamed_total_products_sold = products_sold.sort_values('Total_Products_Sold', ascending=False)
renamed_total_products_sold.head()



#products_sold = reset_combined_sales_df.groupby(['region', 'state', 'city'])
#total_products_sold = products_sold.sum(['units_sold'])
#total_products_sold.sort_values('units_sold')
#renamed_total_products_sold = total_products_sold.rename(columns={'total_sales' : 'Total Sales'})
#renamed_total_products_sold.sort_values('Total Sales', ascending=False)
#renamed_total_products_sold.head()

# Show the top 5 results.
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.

combined_sales_df_pivot = pd.pivot_table(combined_sales_df,
                                                index=['region', 'state', 'city' ],
                                                values='units_sold',
                                                aggfunc='sum'
                                                )
# Rename the "units_sold" column to "Total_Products_Sold"
renamed_products_sold_pivot= combined_sales_df_pivot.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
renamed_products_sold_pivot.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
  # Filter the sales data to get the women's athletic footwear sales data.

womens_data = reindex_combined_sales_df
womens_data = reindex_combined_sales_df.loc[reindex_combined_sales_df['product'].str.contains("Women's Athletic Footwear")]

womens_data
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

womens_ath = womens_data.groupby(['retailer', 'region', 'state', 'city' ])['units_sold'].agg(units_sold = 'sum')
womens_ath = womens_ath.rename(columns = {'units_sold': 'Womens_Footwear_Units_Sold'}).sort_values('Womens_Footwear_Units_Sold', ascending=False)


womens_ath.head()


# Show the top 5 results.
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_ath_pivot = womens_data.pivot_table(
                                        index = ['retailer', 'region', 'state', 'city'],
                                        values = 'units_sold',
                                        aggfunc = 'sum').rename(columns = {'units_sold': 'Womens_Footwear_Units_Sold'}).sort_values(
                                            'Womens_Footwear_Units_Sold', ascending=False)

womens_ath_pivot.head()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"


# Show the top 5 results.
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
day_sales = womens_data.pivot_table(
                            index = 'invoice_date',
                            values = 'total_sales',
                            aggfunc= 'sum').rename({'total_sales': 'Total Sales'}).rename(columns = {'total_sales': 'Total Sales'})
day_sales
# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

resample_df = day_sales.resample('D').sum()
# Sort the resampled pivot table in descending order on "Total Sales".
resample_df.sort_values('Total Sales', ascending=False)

```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

resample_df = day_sales.resample('W').sum()

resample_df.sort_values('Total Sales', ascending=False)

# Sort the resampled pivot table in descending order on "Total Sales".
```
